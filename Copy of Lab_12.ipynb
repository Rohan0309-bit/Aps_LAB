# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP tables
    m = [[0 for _ in range(n)] for _ in range(n)]
    s = [[0 for _ in range(n)] for _ in range(n)] # s[i][j] stores the split point k

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k # Store the split point

    return m, s
```

```python
# Input dimensions
p = [5, 10, 3, 12, 5, 50, 6]

# Call the function to get minimum cost and split points
m_table, s_table = matrix_chain_multiplication(p)

# Get the minimum multiplication cost
min_cost = m_table[1][len(p)-1]
print(f"Minimum multiplication cost: {min_cost}")

# Print optimal parenthesization
def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        k = s[i][j]
        left = print_optimal_parenthesization(s, i, k)
        right = print_optimal_parenthesization(s, k + 1, j)
        return f"({left}{right})"

print("Optimal Parenthesization:", print_optimal_parenthesization(s_table, 1, len(p) - 1))

# Calculate and compare number of multiplications saved
# Naive left-to-right multiplication order: ((((A1A2)A3)A4)...An)
# For matrices A1(p0xp1), A2(p1xp2), ..., An(pn-1xpn)
# Cost of (A1A2) = p[0]*p[1]*p[2]. Result: (p0xp2)
# Cost of ((A1A2)A3) = p[0]*p[2]*p[3]. Result: (p0xp3)
# Total cost for naive: sum(p[0] * p[k] * p[k+1] for k from 1 to num_matrices)

num_matrices = len(p) - 1
naive_cost = 0
if num_matrices > 0:
    current_p_row = p[0]
    for k in range(1, num_matrices + 1):
        # For Ak-1 * Ak, dimensions are current_p_row x p[k] x p[k+1]
        # Simplified: cost of multiplying R_prev * Ak is R_prev_rows * R_prev_cols * Ak_cols
        # In the naive case, R_prev_rows is always p[0]
        # R_prev_cols is p[k]
        # Ak_cols is p[k+1]
        if k == 1: # First multiplication A1*A2
            naive_cost += p[0] * p[1] * p[2]
        else: # Subsequent multiplications (R_prev)*Ak
            naive_cost += p[0] * p[k] * p[k+1]

# A more robust calculation for naive cost (left-to-right)
naive_cost_alt = 0
if num_matrices > 1:
    current_rows = p[0]
    current_cols = p[1]
    for i in range(1, num_matrices): # i from 1 to n-1 (for A1...An)
        # Multiply current result (current_rows x current_cols) with A_{i+1} (p[i] x p[i+1])
        # The current_cols must match p[i]. So current result is p[0] x p[i]
        naive_cost_alt += current_rows * p[i] * p[i+1]

print(f"Naive (left-to-right) multiplication cost: {naive_cost_alt}")
print(f"Number of multiplications saved: {naive_cost_alt - min_cost}")


print("\nDP Table (m_table):")
# Adjust printing for m_table to match problem description (m[i][j] = cost to multiply Ai to Aj)
# m_table has dimensions (n x n). m_table[1][len(p)-1] is the result for A1 to A(n-1)
# The problem description indexes matrices from 1 to n-1 (where n=len(p))
# So the actual table to show for matrices A1 to An-1 would be m_table[1:][1:]
for row_idx in range(1, len(p)):
    print(m_table[row_idx][1:])

print("\nDP Table (s_table - split points):")
for row_idx in range(1, len(p)):
    print(s_table[row_idx][1:])
```
